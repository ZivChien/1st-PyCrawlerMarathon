### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
import csv

# 開啟 CSV 檔案
with open('./data/example.csv', newline='', encoding='UTF-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row)
```

```python
# 1. 取出班次一的每一個時間

'''
Your Code
'''
with open('./data/example.csv', newline='', encoding='UTF-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row[5])
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

'''
Your Code
'''
run = []
with open('./data/example.csv', newline='', encoding='UTF-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    #next(rows)
    for row in rows:
        run.append(row[5])

print(run[:5])
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

'''
Your Code
'''
runs = []
with open('./data/example.csv', newline='', encoding='UTF-8') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    #next(rows)
    for row in rows:
        runs.append(row[5:10])

print(runs[:5])
```
